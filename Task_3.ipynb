```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
#import csv file
data=pd.read_csv('/content/householdtask3.csv')
```

```python
#display top data
display(data.head(10))
```

```python
#Scatter plot with year against own
plt.scatter(data['year'],data['own'])

#adding title to the plot
plt.title("Scatter Plot", fontsize=20,fontweight='bold')

#set x and y label
plt.xlabel("year")
plt.ylabel("own")

#legend
plt.legend(['L(year*own)'])

#show result
plt.show()
```

```python
#line chart with year against own
plt.plot(data['year'],color='green',marker='o', markerfacecolor='red')
plt.plot(data['own'],color='red')

#adding title to the plot
plt.title("Linear Chart")

#set x and y label
plt.xlabel('year')
plt.ylabel('own')

#enable grid
plt.grid(True)

#show result
plt.show()
```

```python
#Bar chart with year against own
plt.bar(data['year'],data['own'],color='red')

#adding title to the plot
plt.title("Bar chart")

#set x and y label
plt.xlabel('year')
plt.ylabel('own')

#show result
plt.show()
```

```python
#Histogram
plt.hist(data['income'],color="pink")

#adding title to the plot
plt.title("Histogram")

#show result
plt.show()
```
